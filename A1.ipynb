**Part 2.1 (10 marks):**

Given $x_i \in \mathrm{R}^{n}$ for $i=1,...,m$ and $y_j \in \mathrm{R}^{n}$ for $j=1,...,k$, where:
- $\mathbf{X} = [x_1,x_2,....,x_m]$ so $\mathbf{X} \in \mathrm{R}^{n \times m}$

- $\mathbf{Y} = [y_1,y_2,....,y_k]$ so $\mathbf{Y} \in \mathrm{R}^{n \times k}$

we define $\mathbf{D} \in \mathrm{R}^{m,k}$ with each element defined as:

$d_{i,j}=\|x_i-y_j\|^2$

Complete the following implementation to complete the matrix $\mathbf{D}$

```python
import numpy as np
import time

# Define the input matrices
X = np.array([[1,2,3,5],
              [5,9,8,6]])  # shape (2, 4) -> 2 features, 4 samples
Y = np.array([[2,0,1],
              [0,4,1]])    # shape (2, 3) -> 2 features, 3 samples

print("X :", X)
print("Y :", Y)

print("X shape:", X.shape)
print("Y shape:", Y.shape)

start_time = time.time()

# Initialize D with shape (m, k) -> (4, 3)
# X.shape[1] is m (columns of X)
# Y.shape[1] is k (columns of Y)
D = np.empty((X.shape[1], Y.shape[1]))

# Loop over each column in X (each sample xi)
for i in range(X.shape[1]):
    xi = X[:, i]
    # Loop over each column in Y (each sample yj)
    for j in range(Y.shape[1]):
        yj = Y[:, j]
        # COMPLETE THE LINE BELOW
        # Calculate squared Euclidean distance: sum of squared differences
        D[i, j] = np.sum((xi - yj)**2)

print("--- %s seconds ---" % (time.time() - start_time))
print("D = \n", D)
```

**Part 2.2 (10 marks):**
1. Show that $d_{i,j}=\sum_{l=1}^n x_{i,l}^2 + \sum_{l=1}^n y_{j,l}^2-2x_i^Ty_j$

2. Verify that the summation $\mathbf{x}_i^T  \mathbf{y}_j$ can be vectorized as the following:
$$
\sum x_i^Ty_j = \mathbf{X}^T\mathbf{Y},
$$

2. Show that each element in the row of the matrix $diag(\mathbf{X}^T\mathbf{X}) \mathbf{1}_{m,k}$ can be expressed as $\sum_{l=1}^n x_{i,l}^2$

where:
- $diag$ is a diagonal matrix (all elements are zeros except those in the diagonal)
- $\mathbf{1}_{m,k}$ is a matrix of ones of dimension $m \times k$


4. Similarly, show that each element in the columns of the matrix $\mathbf{1}_{m,k} diag(\mathbf{Y}^T\mathbf{Y})^T $ can be expressed as $\sum_{l=1}^n y_{j,l}^2$
5. Show that $\mathbf{D}=diag(\mathbf{X}^T\mathbf{X}) \mathbf{1}_{m,k}+\mathbf{1}_{m,k} diag(\mathbf{Y}^T\mathbf{Y})-2\mathbf{X}^T\mathbf{Y}$

Vectorized Derivation of Squared Euclidean Distance1. Scalar ExpansionThe squared Euclidean distance between two vectors $x_i$ and $y_j$ is defined by the sum of squared differences. Expanding the quadratic term $(a - b)^2 = a^2 - 2ab + b^2$ across all components yields:$$d_{i,j} = \sum_{l=1}^{n} (x_{i,l} - y_{j,l})^2 = \sum_{l=1}^{n} x_{i,l}^2 + \sum_{l=1}^{n} y_{j,l}^2 - 2\sum_{l=1}^{n} x_{i,l}y_{j,l}$$Using vector notation where $\|x_i\|^2 = x_i^T x_i$, we simplify this to:$$d_{i,j} = \|x_i\|^2 + \|y_j\|^2 - 2x_i^T y_j$$2. Vectorizing the Dot Product ($2x_i^T y_j$)Let $\mathbf{X}$ and $\mathbf{Y}$ be matrices where columns represent individual data vectors. The matrix product $\mathbf{X}^T \mathbf{Y}$ computes all pairwise dot products simultaneously. Specifically, the entry at $(i, j)$ in the resulting matrix is:$$(\mathbf{X}^T \mathbf{Y})_{i,j} = x_i^T y_j$$3. Vectorizing the Squared Norm of $x$ ($\|x_i\|^2$)To represent the term $\|x_i\|^2$ for every pair $(i, j)$, we use a diagonal matrix of squared norms:$diag(\mathbf{X}^T\mathbf{X})$ extracts the squared norms $\|x_i\|^2$ onto the diagonal.Multiplying by $\mathbf{1}_{m,k}$ (a matrix of ones) broadcasts these values across all $k$ columns.Thus, for every element in row $i$:$$(diag(\mathbf{X}^T\mathbf{X})\mathbf{1}_{m,k})_{i,j} = \|x_i\|^2 = \sum_{l=1}^n x_{i,l}^2$$4. Vectorizing the Squared Norm of $y$ ($\|y_j\|^2$)Similarly, for the $y$ terms, we broadcast the squared norms $\|y_j\|^2$ across all $m$ rows:$diag(\mathbf{Y}^T\mathbf{Y})$ captures the squared norms of each $y_j$.Pre-multiplying by $\mathbf{1}_{m,k}$ ensures that every element in column $j$ contains the norm of $y_j$.$$(1_{m,k} diag(\mathbf{Y}^T\mathbf{Y}))_{i,j} = \|y_j\|^2 = \sum_{l=1}^n y_{j,l}^2$$5. Final Matrix FormBy combining the components from steps 2, 3, and 4, we can express the full distance matrix $\mathbf{D}$ (where $\mathbf{D}_{i,j} = d_{i,j}$) without using any summation loops:$$\mathbf{D} = diag(\mathbf{X}^T\mathbf{X})\mathbf{1}_{m,k} + \mathbf{1}_{m,k}diag(\mathbf{Y}^T\mathbf{Y}) - 2\mathbf{X}^T\mathbf{Y}$$

**Part 2.3 (10 marks):**

Use the result $\mathbf{D}=diag(\mathbf{X}^T\mathbf{X}) \mathbf{1}_{m,k}+\mathbf{1}_{m,k} diag(\mathbf{Y}^T\mathbf{Y})-2\mathbf{X}^T\mathbf{Y}$ to write a vectorized program for better implementation efficiency for the matrix $\mathbf{D}$.

```python
import numpy as np
import time

# Define the input matrices
X = np.array([[1,2,3,5],
              [5,9,8,6]])
Y = np.array([[2,0,1],
              [0,4,1]])

n, m = X.shape
n, k = Y.shape

print("X shape:", X.shape)
print("Y shape:", Y.shape)

start_time = time.time()

# 1. Compute term: -2 * X^T * Y
term_cross = -2 * np.dot(X.T, Y)

# 2. Compute term: diag(X^T * X)
# Sum of squares for each column of X. Result shape (m,)
# Reshape to (m, 1) to act as a column vector for broadcasting
term_X = np.sum(X**2, axis=0).reshape(-1, 1)

# 3. Compute term: diag(Y^T * Y)
# Sum of squares for each column of Y. Result shape (k,)
# Reshape to (1, k) to act as a row vector for broadcasting
term_Y = np.sum(Y**2, axis=0).reshape(1, -1)

# 4. Combine terms
# Broadcasting automatically handles the "multiplication by ones matrix" concept
D = term_X + term_Y + term_cross

print("--- %s seconds ---" % (time.time() - start_time))
print(D)
```

```python
import numpy as np
import time

# Define the input matrices
X = np.array([[1,2,3,5],
              [5,9,8,6]])
Y = np.array([[2,0,1],
              [0,4,1]])

n, m = X.shape
n, k = Y.shape

print("X shape:", X.shape)
print("Y shape:", Y.shape)

start_time = time.time()

# 1. Compute the squared L2 norms for each vector in X and Y
# Sum over axis=0 (features) to get a value for each sample
# X**2 is element-wise square
sx = np.sum(X**2, axis=0)  # Shape: (m,)
sy = np.sum(Y**2, axis=0)  # Shape: (k,)

# 2. Compute the cross term (dot product)
# -2 * X.T @ Y -> Shape: (m, k)
d_cross = -2 * np.dot(X.T, Y)

# 3. Use Broadcasting to compute D
# sx[:, np.newaxis] converts shape (m,) -> (m, 1)
# sy[np.newaxis, :] converts shape (k,) -> (1, k) (optional, implicit in broadcasting)
# The addition automatically broadcasts these to (m, k)
D = sx[:, np.newaxis] + sy[np.newaxis, :] + d_cross

print("--- %s seconds ---" % (time.time() - start_time))
print(D)
```

### **Q3** (20 marks)  **Gradient Descent** ###

**Part 3.1 (10 marks):**
Given the matrix $\mathbf{A} \in \mathbb{R}^{m \times n}$ and the output vector $\mathbf{y} \in \mathbb{R}^m$, in order to solve the following linear equation:
$$
\mathbf{A} \mathbf{x} = \mathbf{y}
$$
for the input vector $\mathbf{x} \in \mathbb{R}^n$, we may seek to solve the following optimization problem:
$$
\min_{\mathbf{x}}  \big\Vert \mathbf{A} \mathbf{x} - \mathbf{y} \big\Vert^2
$$

Solving this optimization problem is equivallent to solving the original linear equation. Derive a gradient decent algorithm to solve this optimization optimization.

To solve the linear equation $\mathbf{Ax} = \mathbf{y}$, we minimize the objective function $J(\mathbf{x}) = \|\mathbf{Ax} - \mathbf{y}\|^2$,which represents the squared Euclidean norm of the error.
Here is the derivation for the gradient descent algorithm:Define the Objective Function:
 The goal is to minimize the cost function $J(\mathbf{x})$, defined as the squared norm of the residual vector.
 Expand the Objective Function: By applying the property $\|\mathbf{z}\|^2 = \mathbf{z}^T\mathbf{z}$, we expand the expression into a quadratic form:$J(\mathbf{x}) = (\mathbf{Ax} - \mathbf{y})^T(\mathbf{Ax} - \mathbf{y})$$J(\mathbf{x}) = \mathbf{x}^T\mathbf{A}^T\mathbf{Ax} - \mathbf{x}^T\mathbf{A}^T\mathbf{y} - \mathbf{y}^T\mathbf{Ax} + \mathbf{y}^T\mathbf{y}$Because the term $\mathbf{x}^T\mathbf{A}^T\mathbf{y}$ is a scalar, it is equivalent to its transpose $\mathbf{y}^T\mathbf{Ax}$.
 This simplifies the function to $J(\mathbf{x}) = \mathbf{x}^T\mathbf{A}^T\mathbf{Ax} - 2\mathbf{x}^T\mathbf{A}^T\mathbf{y} + \mathbf{y}^T\mathbf{y}$.Compute the Gradient: To find the direction of steepest ascent, we take the gradient of $J(\mathbf{x})$ with respect to $\mathbf{x}$ using matrix calculus identities:$\nabla_{\mathbf{x}}J(\mathbf{x}) = 2\mathbf{A}^T\mathbf{Ax} - 2\mathbf{A}^T\mathbf{y}$Factoring out the shared terms, we get $\nabla_{\mathbf{x}}J(\mathbf{x}) = 2\mathbf{A}^T(\mathbf{Ax} - \mathbf{y})$.Update Rule: The gradient descent algorithm iteratively updates the vector $\mathbf{x}$ by taking steps in the opposite direction of the gradient. Using $\eta$ (eta) as the learning rate, the update rule for iteration $t+1$ is:$\mathbf{x}^{(t+1)} = \mathbf{x}^{(t)} - \eta \nabla_{\mathbf{x}}J(\mathbf{x}^{(t)})$By substituting our derived gradient: $\mathbf{x}^{(t+1)} = \mathbf{x}^{(t)} - \eta \cdot 2\mathbf{A}^T(\mathbf{Ax}^{(t)} - \mathbf{y})$This can be written as: $\mathbf{x}^{(t+1)} = \mathbf{x}^{(t)} - 2\eta\mathbf{A}^T(\mathbf{Ax}^{(t)} - \mathbf{y})$.

```python
# for part 3.2, write your code here

import jax.numpy as jnp
from jax import grad

# 1. Define the objective function: J(x) = ||Ax - y||^2
def loss_fn(x, A, y):
    diff = jnp.dot(A, x) - y
    return jnp.sum(diff**2)

# 2. Use JAX to automatically create the gradient function
# grad(loss_fn) creates a function that computes dJ/dx
grad_loss = grad(loss_fn)

# 3. Define the Gradient Descent Solver
def solve_gradient_descent(A, y, learning_rate=0.01, n_iter=5000):
    # n is the number of features (columns in A)
    n = A.shape[1]

    # Initialize x (e.g., with zeros)
    x = jnp.zeros((n, 1))

    # Gradient Descent Loop
    for i in range(n_iter):
        # Calculate gradient using auto-grad
        gradients = grad_loss(x, A, y)

        # Update rule: x = x - learning_rate * gradient
        x = x - learning_rate * gradients

    return x

# --- Solve Case 1 ---
print("--- Solving Case 1 ---")
x1_star = solve_gradient_descent(A1, y1, learning_rate=0.01)
print("Found solution x* for Case 1:\n", x1_star)

# --- Solve Case 2 ---
print("\n--- Solving Case 2 ---")
# Case 2 may require slightly different tuning
x2_star = solve_gradient_descent(A2, y2, learning_rate=0.005)
print("Found solution x* for Case 2:\n", x2_star)
```
